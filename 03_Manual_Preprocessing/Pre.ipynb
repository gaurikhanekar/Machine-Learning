# Dataframes Fillna

```python
df["AirBags"].isna().sum()
```

```python
mode = df["AirBags"].mode()[0]
df["AirBags"] = df["AirBags"].fillna(mode)
```

```python
df["AirBags"].isna().sum()
```

```python
for i in df.columns:
    print(i)
```

```python
df.isna().sum()
```

```python
def replacer(df):
    cat_cols = df.select_dtypes(include= "object").columns
    num_cols = df.select_dtypes(include= "number").columns
    for i in df.columns:
        if i in cat_cols:
             mode = df[i].mode()[0]
             df[i] = df[i].fillna(mode)
        elif i in num_cols:
            mean = df[i].mean()
            df[i] = df[i].fillna(mean)
     
```

```python
df.isna().sum()
```

```python
replacer(df)
```

```python
df.isna().sum()
```

```python
df.shape
```

```python
a = df.select_dtypes(include= "object").nunique()
a
```

```python
card = a / len(df)
print(card)
```

```python
card[card >= 0.9]
```

```python
high_card = card[card >= 0.9].index
high_card
```

```python
df = df.drop(columns= high_card)
```

```python
df.head()
```

# step 3 = seperate X and y (Weight)

```python
X = df.drop(columns= ["id", "Weight"])
X.head()
```

```python
Y = df["Weight"]
Y.head()
```

# Apply preprocessing on X

```python
X_cat = X.select_dtypes(include= "object")
X_cat
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(
    handle_unknown= "ignore", sparse_output= False, drop="first"
).set_output(transform= "pandas")
ohe.fit(X_cat)

```

```python
X_cat_pre = ohe.transform(X_cat)
X_cat_pre
```

```python
X_num = X.select_dtypes(include = "number")
X_num
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler().set_output(transform= "pandas")
scaler.fit(X_num)
```

```python
X_num_pre = scaler.transform(X_num)
X_num_pre.head()
```

```python
X_cat_pre.head()
```

```python
X_pre = X_num_pre.join(X_cat_pre)
X_pre.head()
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split

xtrain, xtest, ytrain, ytest = train_test_split(X_pre, Y, test_size=0.2, random_state=21)
```

# Build the Model

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(xtrain, ytrain)
```

# Model Evaluation

```python
# R2 score on train
model.score(xtrain, ytrain)
```

```python
# R2 score on test
model.score(xtest, ytest)
```

```python
from sklearn.metrics import(
    root_mean_squared_error,
    mean_absolute_error,
    mean_absolute_percentage_error,
    r2_score
)
```

```python
def evaluate_model(model, x, y):
    ypred = model.predict(x)
    rmse = root_mean_squared_error(y, ypred)
    mae = mean_absolute_error(y, ypred)
    mape = mean_absolute_percentage_error(y, ypred)
    r2 = r2_score(y, ypred)
    print(f"RMSE : {rmse:.2f}")
    print(f"MAE : {mae :.2f}")
    print(f"MAPE : {mape:.2%}")
    print(f"R2 : {r2:.2%}")
```

```python
print("Train Results : ")
evaluate_model(model, xtrain, ytrain)
```

```python
print("Train Results : ")
evaluate_model(model, xtest, ytest)
```
